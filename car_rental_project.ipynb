```python
# Importing the necessary classes
from car_rental import CarRental
from customer import Customer

def main():
    # Create an instance of CarRental with 100 cars
    rental_shop = CarRental(100)
    customer = Customer()

    while True:
        print("""
        ====== Car Rental Shop =======
        1. Display available cars
        2. Request a car on an hourly basis $5
        3. Request a car on a daily basis $20
        4. Request a car on a weekly basis $60
        5. Return a car
        6. Exit
        """)
        try:
            choice = int(input("Enter your choice: "))
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 6.")
            continue

        if choice == 1:
            rental_shop.display_stock()

        elif choice == 2:
            cars = customer.request_car()
            if cars != -1:
                rental_time = rental_shop.rent_hourly(cars)
                customer.rental_basis = 1
                customer.rental_time = rental_time

        elif choice == 3:
            cars = customer.request_car()
            if cars != -1:
                rental_time = rental_shop.rent_daily(cars)
                customer.rental_basis = 2
                customer.rental_time = rental_time

        elif choice == 4:
            cars = customer.request_car()
            if cars != -1:
                rental_time = rental_shop.rent_weekly(cars)
                customer.rental_basis = 3
                customer.rental_time = rental_time

        elif choice == 5:
            request = customer.return_car()
            rental_shop.return_car(request)
            customer.rental_basis, customer.rental_time, customer.cars = 0, 0, 0

        elif choice == 6:
            print("Thank you for using our service!")
            break

        else:
            print("Invalid input. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
